# Datos espaciales
Un dato espacial es aquel que puede ser asociado a una posición en el espacio. Ya sea en un plano 2D, 3D o n-dimensional

```python
import matplotlib.pyplot as plt

#Simular un conjunto de datos que representen la elevacion puntual en un terreno

#Crear un conjunto de datos de prueba
elevacion = np.random.randint(0,3000, size=(100,100))
elevacion
```

```python
plt.imshow(elevacion, cmap="terrain")
plt.colorbar(label="Altura(m)")
plt.show()
```

# Crear un ndarray de dos dimensiones con la estatura de los estudiantes

```python
import numpy as np
import matplotlib.pyplot as plt

#crear un ndarray de la estatura de los estudiantes 
estaturas = np.array ([[1.88,1.60, 1.60],[1.60, 1.65,1.59],[1.75,1.54,1.60],[1.72,1.74,1.80],[1.62,1.60,1.60],[1.65,1.64,1.70]])
```

```python
#calcular la media de las estaturas
media = np.mean(estaturas)
media
```

```python
#calcular la mediana de las estaturas
mediana = np.median(estaturas)
mediana
```

```python
#calcular la desviacion estandar de las estaturas
desviacion= np.std(estaturas)
desviacion
```

```python
elevacion = np.random.randint(0,200, size=(100,100))
elevacion 
plt.imshow(elevacion, cmap="terrain")
plt.colorbar(label="Estaturas (cm)")
plt.show()
```
